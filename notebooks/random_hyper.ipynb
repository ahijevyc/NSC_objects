```python
%%time
import numpy as np
import pdb
import random

param_grid = {
    'neurons':   [ 16, 64, 256, 1024 ],
    'layers':    [ 1, 2 ],
    'optimizer':     [ 'SGD', 'Adam'],
    'learning_rate': [ 0.1, 0.01, 0.001 ],
    'dropout':       [ 0, 0.1, 0.2 ],
    'reg_penalty':   [ 0, 0.1, 0.01 ],
    'epochs':        [ 10, 30 ],
    'batchnorm':    [ True, False ]
}

num_combos = 1
for v in param_grid.values():
    num_combos *= len(v)
print(num_combos, 'combinations')

num_samples = 1000
random.seed(10)
config_dict = {}
config_strings = []

while len(config_strings) < num_samples:
    # Randomly sample from dictionary
    config_num = len(config_strings)
    random_params = {k: random.sample(v, 1)[0] for k, v in param_grid.items() }
    config_string = "--glm "
    config_string += " ".join([f"--{k} {v}" for k, v in random_params.items()])
    config_string += " --trainstart 19000101 --trainend 20201202 --teststart 20201202T12 --testend 20220101 --model HRRR"
    
    config_string = config_string.replace("--batchnorm True","--batchnorm").replace(" --batchnorm False","")
    
    
    # make sure this dictionary doesn't already exist
    if config_string not in config_strings:
          config_dict[config_num] = random_params
          config_strings.append(config_string)
    
print("\n".join(config_strings[250:500]))
```
