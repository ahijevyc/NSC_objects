```python
import os

import geopandas as gpd
from shapely.geometry import Polygon
tmpdir = os.getenv("TMPDIR")
s = "blockTX"
polygon = Polygon(
    [
        (-98, 26),
        (-102, 30),
        (-103, 29),
        (-106, 32),
        (-103, 32),
        (-103, 36.5),
        (-100, 36.5),
        (-100, 34),
        (-94, 33),
        (-94, 30),
        (-96, 29),
    ]
)

coords_gdf = gpd.GeoDataFrame(index=[0], geometry=[polygon], crs="EPSG:4326")
coords_gdf.plot(color="brown")
ofile = tmpdir + f"{s}.shp"
coords_gdf.to_file(ofile)
```
